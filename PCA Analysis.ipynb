```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
np.set_printoptions(precision=3,suppress=True)
data = pd.read_csv(&quot;Stroke.csv&quot;)
data.fillna(data.mean(), inplace=True)
data = data.drop([&#39;Age Group&#39;, &#39;stroke&#39;], axis=1)
labelEncoder = LabelEncoder()
labelEncoder.fit(data[&#39;gender&#39;])
data[&#39;gender&#39;] = labelEncoder.transform(data[&#39;gender&#39;])
labelEncoder = LabelEncoder()
labelEncoder.fit(data[&#39;ever_married&#39;])
data[&#39;ever_married&#39;] = labelEncoder.transform(data[&#39;ever_married&#39;])
labelEncoder = LabelEncoder()
labelEncoder.fit(data[&#39;work_type&#39;])
data[&#39;work_type&#39;] = labelEncoder.transform(data[&#39;work_type&#39;])
labelEncoder = LabelEncoder()
labelEncoder.fit(data[&#39;Residence_type&#39;])
data[&#39;Residence_type&#39;] = labelEncoder.transform(data[&#39;Residence_type&#39;])
labelEncoder = LabelEncoder()
labelEncoder.fit(data[&#39;smoking_status&#39;])
data[&#39;smoking_status&#39;] = labelEncoder.transform(data[&#39;smoking_status&#39;])
scaler = StandardScaler()
scaler.fit(data)
normalData = scaler.transform(data)
pca = PCA()
transformedPCA = pca.fit_transform(normalData)
print(&quot;Explained variance:&quot;, pca.explained_variance_)

26

print(&quot;Explained variance ratio:&quot;, pca.explained_variance_ratio_)
print(&quot;Cumulative explained variance ratio:&quot;, pca.explained_variance_ratio_.cumsum())
print()
print(&quot;Feature Weights&quot;)
for i in data.columns:
print(i, end=&#39; &#39;)
print()
print(abs( pca.components_ ))
weights = pd.DataFrame(abs(pca.components_))
weights.columns = data.columns
writer = pd.ExcelWriter(&#39;Feature Weights for Stroke.xlsx&#39;)
weights.to_excel(writer,index=False)
writer.save()

print()
print(&#39;###############################################&#39;)
print()

data = data.drop([&#39;ever_married&#39;,&#39;work_type&#39;,&#39;bmi&#39;], axis=1)

scaler = StandardScaler()
scaler.fit(data)
normalData = scaler.transform(data)
pca = PCA()
transformedPCA = pca.fit_transform(normalData)
print(&quot;Explained variance:&quot;, pca.explained_variance_)
print(&quot;Explained variance ratio:&quot;, pca.explained_variance_ratio_)
print(&quot;Cumulative explained variance ratio:&quot;, pca.explained_variance_ratio_.cumsum())
print()
print(&quot;Feature Weights&quot;)

for i in data.columns:
print(i, end=&#39; &#39;)
print()

27

print(abs( pca.components_ ))
weights2 = pd.DataFrame(abs(pca.components_))
weights2.columns = data.columns
writer2 = pd.ExcelWriter(&#39;Feature Weights for Stroke2.xlsx&#39;)
weights2.to_excel(writer2,index=False)
writer2.save()
```
